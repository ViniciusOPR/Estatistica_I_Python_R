Testes de Normalidade com Python

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
#Criação de uma variável com dados em uma distribuição normal com a função RVS (100 elementos)
dados = norm.rvs(size=100)
dados
```

```python
#Histograma
plt.hist(dados, bins=20)
plt.title('Dados')
```

```python
#Geração de gráfico (qqplot) para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

```python
#Execução do teste de Shapiro - Segundo argumento é o valor de P, não há como rejeitar a hipótese nula
stats.shapiro(dados)
```

Dados não normais

```python
dados2 = skewnorm.rvs(4, size=1000)
```

```python
#histograma
plt.hist(dados2, bins = 20)
plt.title('Dados')
```

```python
# Geração de gráfico para verificar se a distribuição é normal
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True,   plot=ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
